```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/FDE/customers.csv")
```

```python
print(df)
```

#Removing Reduntant Rows

```python
new_df = df.drop_duplicates(keep = "first")
```

```python
print(new_df)
```

#Handling Missing Values : NULL City values = "Unknown"

```python
new_df['city'] = new_df['city'].fillna('unknown')
print(new_df)
```

# Change a single column name.




```python
new_df.columns = new_df.columns.str.upper()
print(new_df)
```

#New Column with Functions

```python
print(new_df)
```

```python
new_df = pd.DataFrame(new_df)
```

```python
conditions = [
    new_df['ANNUAL_SPEND'] < 80000,
    (new_df['ANNUAL_SPEND'] >= 80000) & (new_df['ANNUAL_SPEND'] < 120000)
]
choices = ['low', 'medium']
new_df['SPEND_CATEGORY'] = np.select(conditions, choices, default='high')
```

```python
grouped_data = new_df.groupby(['CITY', 'SPEND_CATEGORY']).agg(
    total_customers=('CUSTOMER_ID', 'count'),
    average_annual_spend=('ANNUAL_SPEND', 'mean')
)
print(grouped_data)
```

```python
new_df.to_csv('etl_analytics_ready.csv', index=False)
print('DataFrame saved to etl_analytics_ready.csv')
```
